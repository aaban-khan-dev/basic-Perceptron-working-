```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame();
and_data = pd.DataFrame();
xor_data = pd.DataFrame();
```

```python
or_data['input1'] = [0,0,1,1]
or_data['input2'] = [0,1,0,1]
or_data['output'] = [0,1,1,1]
```

```python
and_data['input1'] = [0,0,1,1]
and_data['input2'] = [0,1,0,1]
and_data['output'] = [0,0,0,1]
```

```python
xor_data['input1'] = [0,0,1,1]
xor_data['input2'] = [0,1,0,1]
xor_data['output'] = [0,1,1,0]
```

```python
sns.scatterplot(x = or_data['input1'],y = or_data['input2'],hue = or_data['output'],s = 200)
```

```python
sns.scatterplot(x = and_data['input1'],y = and_data['input2'],hue = and_data['output'],s = 200)
```

```python
sns.scatterplot(x = xor_data['input1'],y = xor_data['input2'],hue = xor_data['output'],s = 200)
```

```python
from sklearn.linear_model import Perceptron
```

```python
or_classifier = Perceptron()
and_classifier = Perceptron()
xor_classifier = Perceptron()
```

```python
or_classifier.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
and_classifier.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
xor_classifier.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
or_classifier.coef_
```

```python
or_classifier.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions

# setup plots for side by side comparisons
fig,axes = plt.subplots(1,3,figsize=(18,5))

#AND classifier Decision boundary
axes[0].set_title('AND classifier')
plot_decision_regions(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values,clf = and_classifier,legend = 2,ax = axes[0])

#OR classifier Decision boundary
axes[1].set_title('OR classifier')
plot_decision_regions(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values,clf = or_classifier,legend = 2,ax = axes[1])

#XOR classifier Decision boundary
axes[2].set_title('XOR classifier')
plot_decision_regions(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values,clf = xor_classifier,legend = 2,ax = axes[2])
```
